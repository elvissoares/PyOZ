```python
import sys
sys.path.insert(0, '../src/')
from hankel import hankel_forward, hankel_inverse
import torch 

n = 512
dr = 0.02

device = torch.device(
    "cuda" if torch.cuda.is_available() else "cpu"
)

dtype = torch.float64

i = torch.arange(n, device=device, dtype=dtype)

dk = torch.pi / (n * dr)

r = (i + 0.5) * dr
k = (i + 0.5) * dk
```

```python
fr = torch.exp(-r**2)

fk = hankel_forward(fr, r, k, dr)
fr_recovered = hankel_inverse(fk, r, k, dk)

error = torch.max(torch.abs(fr_recovered - fr))

print(f"Erro máximo: {error.item():.8e}")
```
